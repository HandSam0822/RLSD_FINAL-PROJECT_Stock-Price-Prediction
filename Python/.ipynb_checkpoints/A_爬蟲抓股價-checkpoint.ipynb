```python
# step1. import package 
import requests
import pandas as pd
import numpy as np
from io import StringIO
df2 = pd.DataFrame()
# step2. 進入目標網站,爬取盤後資訊 

date = ['20180330','20180430','20180531','20180629','20180731',
     '20180831','20180928','20181031','20181130','20181231',
     '20190131','20190131','20190131','20190131','20190131',
     '20190131','20190228','20190329','20190430','20190531',
     '20190628','20190731','20190830','20190930','20191031']
     


for date in date:
    try:
        r = requests.post('https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=csv&date=' + date + '&stockNo=3008')

        # step3. 篩選出個股盤後資訊
        str_list = []
        for i in r.text.split('\n'):
            if len(i.split('",')) == 10 and i[0] != '=':       
                i = i.strip(",\r\n")
                str_list.append(i)  

        df = pd.read_csv(StringIO("\n".join(str_list)))  
        df2 = pd.concat([df2,df],axis = 0)
        #step4. 印出選股資訊
        print(df2.shape)

    except Exception as e:
        print(f'An error occured:{e}')

df2.to_csv('/Users/liaoshousan/Desktop/3008_大立光.csv',index = None)
            #     pd.set_option('display.max_rows', None)
        #     df.head(150)
%reset -f
```
